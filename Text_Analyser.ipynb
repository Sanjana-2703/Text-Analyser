```python
# setting up the environment 
import requests
from bs4 import BeautifulSoup
```

```python
url = ('https://en.wikipedia.org/wiki/Indian_Premier_League')
r = requests.get(url)
htmlcontent = r.content
soup = BeautifulSoup(htmlcontent,'lxml')
print(soup.prettify)
```

```python
# Title of the web Page
title =  soup.title
print(title.string)
```

```python
# sentence count
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
```

```python
text = ""
for pr in soup.find_all('p'):
  text +=pr.text
text = re.sub(r'\[[0-9]*\]',' ',text)
# text = re.sub(r's+',' ',text)
text = text.lower()
# text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
print(text)
```

```python
sentences  = nltk.sent_tokenize(text)
for i in sentences:
  print(i)
```

```python
# no. of sentence
print("no. of sentences is = ",len(sentences))
```

```python
# words count 
words =  text.split()
print(words)
words.sort()
for i in set(words):
  lf = text.count(i)
  print(i,":",lf)
```

```python
# Winning Team Names 
table = soup.find('table',class_="wikitable plainrowheaders")
tr =  table.find_all('tr')
rows = table.find_all("tr") 
l=[]
for row in rows[2:]:
    cells = row.find_all(["td", "th"])
    cells_text = [cell.get_text(strip=True) for cell in cells][:3]
    l.append(cells_text)
for i in range(len(l)):
  print(l[i][0].split('D')[0] + " -----  ",end='',sep='' )
  print(l[i][2].split('[')[0])
```

```python
#IPL website in different languages 
b = soup.find_all("a",class_='interlanguage-link-target')
for i in b:
  x=i.get('title')
  print(x, end = " ", sep= "---")
  y=i.get('href')
  print(y)
```
